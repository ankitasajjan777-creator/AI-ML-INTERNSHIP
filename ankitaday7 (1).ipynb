```python
def mean(array):
  n = len(array)
  return np.sum(array)/n
```

```python
mean(np.array([1,2,3,4,5,6]))
```

```python
def variance(array):
  n = len(array)
  m = mean(array)
  return np.sum((array - m)**2)/n
```

```python
variance(np.array([1,2,3,4,5,6]))
```

```python
##min-max scalling using numpy
```

```python
def min_max_scaler(array):
    array = np.array(array)
    min_val = np.min(array)
    max_val = np.max(array)
    scaled = (array - min_val) / (max_val - min_val)
    return scaled
```

```python
min_max_scaler(np.array([1,2,3,4,5,6]))
```

```python
##standard  scaler
```

```python
def standard_scaler(array):
    array = np.array(array)
    mean_val = np.mean(array)
    std_val = np.std(array)
    scaled = (array - mean_val) / std_val
    return scaled
```

```python
standard_scaler(np.array([1,2,3,4,5,6]))
```

```python
#sigmoid function
```

```python
def sigmoid(array):
    array = np.array(array)
    return 1 / (1 + np.exp(-array))
```

```python
sigmoid(np.array([0, 1, -1, 2]))
```

```python
# Relu function
```

```python
def relu(array):
    array = np.array(array)
    return np.maximum(0, array)
```

```python
relu(np.array([-2, -1, 0, 3, 5]))
```

```python
#leaky_relu function
```

```python
def leaky_relu(array, alpha=0.01):
    array = np.array(array)
    return np.where(array > 0, array, alpha * array)
```

```python
leaky_relu(np.array([-5, -1, 0, 2, 4]))
```

```python
# softmax function
```

```python
def softmax(array):
    array = np.array(array)
    exp_vals = np.exp(array - np.max(array))   # for numerical stability
    return exp_vals / np.sum(exp_vals)
```

```python
softmax(np.array([2, 1, 0]))
```

```python
# sigmoid deveration
```

```python
def sigmoid_derivative(array):
    array = np.array(array)
    sig = 1 / (1 + np.exp(-array))   # sigmoid
    return sig * (1 - sig)
```

```python
sigmoid_derivative(np.array([0, 1, -1]))
```
