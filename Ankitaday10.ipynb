```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
data = pd.read_csv("sample2.csv")
data
```

```python
data.head()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt

fig , axs = plt.subplots(1,4,figsize=(16,6))
axs[0].boxplot(data.Age)
axs[1].boxplot(data.Salary)
axs[2].boxplot(data.Experience)
axs[3].boxplot(data.Monthly_Sales)

plt.show()
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
```

```python
IQR
```

```python
upper_limit = Q3 + 1.5 * IQR
lower_limit = Q1 - 1.5 * IQR
upper_limit, lower_limit
```

```python
if 'Unnamed: 0' in data.columns:
    data_for_outliers_check = data.drop(columns=['Unnamed: 0'])
else:
    data_for_outliers_check = data.copy()

outliers = ((data_for_outliers_check < lower_limit) | (data_for_outliers_check > upper_limit)).any(axis = 1)
outliers_data = data[outliers]
cleaned_data = data[~outliers]
cleaned_data.head()
```

```python
cleaned_data = data[~outliers]
fig , axs = plt.subplots(1,4,figsize=(16,6))
axs[0].boxplot(cleaned_data.Age)
axs[1].boxplot(cleaned_data.Salary)
axs[2].boxplot(cleaned_data.Experience)
axs[3].boxplot(cleaned_data.Monthly_Sales)

plt.show()
```
